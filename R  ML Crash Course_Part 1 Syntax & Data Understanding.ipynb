### List
Lists provide a group of named items, not unlike a map.
- You can define a new list with the list() function. 
- A list can be initialized with values or empty

```r
# create a list of name items
a <- list(aa=1, bb=2, cc=3)
a
```

#### Note that the named values in the list can be accessed using the dollar operator ($).

```r
# add a named item to list
a$dd = 4
a
```

### Matrices
- A matrix is a table of data of the same type (e.g. numeric). 
- It has dimensions (rows and
columns) and
- the columns can be named

```r
data <- c(1, 2, 3, 4, 5, 6)
headings <- list(NULL, c("a","b","c"))
m <- matrix(data, nrow=2, ncol=3, byrow=TRUE, dimnames=headings)
m
```

- A lot of useful plotting and machine learning algorithms require the data to be provided as
a matrix. 
- Note the syntax to index into rows [1,] and columns [,1] of a matrix.

```r
m [1,]
```

```r
m[,1]
```

## Data Frame
- Data frames are useful for representing tables of your data in R
- A matrix is a much simpler structure, intended for mathematical operations. 
- A data frame
is more suited to representing a table of data and 
- is expected by modern implementations of
machine learning algorithms in R

```r
# create a new data frame
years <- c(1980, 1985, 1990)
scores <- c(34, 44, 83)
df <- data.frame(years, scores)
df
```

#### you can index into rows and columns of a data frame just like you can for a matrix.

```r
df[1,]
```

```r
df[,1]
```

#### you can reference a column using its name just like a list (df$years)

```r
df$years
```

## Flow Control
### IF-THEN-ELSE

```r
a <- 66
if (a > 55) {
print("a is more than 55")
} else {
print("A is less than or equal to 55")
}
```

### For Loop

```r
# for loop
mylist <- c(55, 66, 77, 88, 99)
for (value in mylist) {
print(value)
}
```

### While Loop

```r
# while loop
a <- 100
while (a < 500) {
a <- a + 100
}
a
```

## Class Distribution
- In a classication problem you must know the proportion of instances that belong to each class
label. 
- This is important because it may highlight an imbalance in the data, that if severe may
need to be addressed with rebalancing techniques.

```r
y <- df1$PRODUCT_COLOR_CODE
cbind(freq=table(y), percentage=prop.table(table(y))*100)
```
